```python
a=12
b=12
c=a+b
print(c)
c=a-b
print(c)
c=a*b
print(c)
c=a/b
print(c)
c=a**b
print(c)
```

列表

```python
players = ['charles', 'martina', 'michael', 'florence', 'eli'] 
print(players[0:3]) #players[0:3]即为切片，选中开头与结尾的元素位置
print(players[-3])#选中列表中的倒数第3个元素
print(players[:3])#选中列表的前3个元素
print(players[-3:])#选中列表的后3个元素

```

循环

```python
magicians = ['alice', 'david', 'carolina'] 
for magician in magicians:   #将magicians中的元素依次打印出来
  print(magician) 
  
```
